**.何回折ったら富士山を超えるか**

「Pythonを使ってみよう」の課題では「1枚の紙を43回折り曲げた時の厚さを計算するプログラム」を実装し、

43回折れば月に届くことを確認しました。

では、「何回折ったら目的の厚さになるのか」を計算するプログラムを実装しましょう。

紙を折った時の厚さの数式
紙を $n$ 回折ったときの厚さを $t_{n}$ としたとき、以下のように表せます。


\[t_{n} = t_{0}×2^{n}\]
一般的なコピー用紙の厚さはおよそ 0.00008m なので、$t_{0} = 0.00008$ とします。  

## 【問題1】何回折ったら富士山を超えるか

厚さが「富士山の高さ(3776m)」を超えるのに必要な紙を折る最小回数を計算するプログラムを作成してください。

```python
# 富士山の高さ
fujisan = 3776
# 紙の厚さ
THICKNESS = 0.00008
# 折った回数（初期値は０）
count = 0
# 合計の厚さを計算する用にもう一つ用意
total_height = THICKNESS
# 富士山の高さに届くまで繰り返す処理
while total_height <= fujisan:
    # 折る回数の倍数を厚さに掛ける計算
    total_height = THICKNESS * (2 ** count)
    count += 1

    
# 回答
print('厚さ{}のコピー用紙を、合計で{}回折ると、富士山の高さ３７７６mを超えます。'.format(THICKNESS, count))
    
```

## 【問題2】任意の厚さに対応した関数
高さを入力すると、それを超えるのに必要な紙を折る最小回数を出力する関数を実装してください。

引数として、折る前の厚さ t0も設定できるようにしましょう。


この関数を利用して、「最も近い太陽以外の恒星」に到達するには紙を何回折る必要があるか計算してください。

「最も近い太陽以外の恒星」は「プロキシマ・ケンタウリ」で、地球からの距離はおよそ 
4.0175

m です。

```python
def fold_num(height = 0.00008):
    
    ''' 高さを入力して、その高さを超えるのに必要な紙を折る回数を出力する関数
    Parameteres
    -----------
    height: int
    任意の高さ（初期値は紙のサイズである０.０００８m
    
    count
    回数の初期値
    
    total_height
    この中に長さを倍がけして出てくる数字を代入する
    
    Returns
    count
    任意の高さを超えるまでの回数を戻り値として返す
    -------
    
    '''
    # 紙の厚さ（メートル）
    THICKNESS = 0.00008

    count = 0

    total_height = THICKNESS

    while total_height <= height:

        total_height = THICKNESS * (2 ** count)

        count += 1
    
    return count

# 回答①
count = fold_num(1000)
print("入力された高さまで到達するには{}回、紙を折る必要があります".format(count))

# 回答②
Proxima_Centauri = 4.0175 * (10 ** 16)
print("「最も近い太陽以外の恒星」は「プロキシマ・ケンタウリ」で、その距離は{}です！".format(math.floor(Proxima_Centauri)))
    
```

## 【問題3】必要な紙の長さ
実際のところ身近にある紙は43回はおろか10回も折れません。

しかし、大きな紙ほどたくさん折ることができ、トイレットペーパーのような長い紙を

一方向に折ることを考えた時の折れる回数を求める公式が存在します。


厚さ $t_{0}$ の紙をn回折るために必要な紙の長さLは以下の公式で求まります。


\[L = \frac{\pi t_{0}}{6}(2^n + 4)(2^n - 1)\]


この公式を利用して、** 実際に紙を折り任意の厚さに到達するために必要な紙の長さを求める関数を実装してください。 **



そしてこれを使って「月」「富士山」「最も近い太陽以外の恒星」に届くために必要な紙の長さを求めてください。

```python
import math

def fold_num(height_len):
    
    ''' 任意の高さを入力して、その高さを超えるのに必要な紙の長さを計算する関数
    Parameteres
    -----------
    height_len :int
        任意の高さを入力する
    
    Returns
        total_len
            必要となるトータルの紙の長さを戻り値で返す
        # 紙の厚さ（メートル）
        
    while total_height <= height_len:
        # 任意の高さにたどり着くまで、紙を倍にかけた数字をtotal_heightの変数に代入する
        total_height = THICKNESS * (2 ** count)
        
        # 任意の高さに辿り着くまで、その回数を数える
        count += 1
        # 任意の高さに辿り着くまで、トータルの長さを数える
        total_len += total_height 
    -------
    
    '''
    # 紙の厚さ（メートル）
    THICKNESS = 0.00008

    count = 0

    total_height = THICKNESS
    total_len = 0

    while total_height <= height_len:
        # 任意の高さにたどり着くまで、紙を倍にかけた数字をtotal_heightの変数に代入する
        total_height = THICKNESS * (2 ** count)
        
        # 任意の高さに辿り着くまで、その回数を数える
        count += 1
        # 任意の高さに辿り着くまで、トータルの長さを数える
        total_len += total_height 
    
#  戻り値としてトータルの長さを返す
    return total_len

# 回答①
print("入力された高さまで届くには{:.2f}メートルの紙が必要です".format(fold_num(1000)))

moon = 384400
fujisan = 3776
Proxima_Centauri = 4.0175 * (10 ** 16)

# 回答②
print("月の高さまで届くには{}メートルの紙が必要です".format(fold_num(moon)))
print("富士山山頂の高さまで届くには{}メートルの紙が必要です".format(fold_num(fujisan)))
print("プロキシマケンタウリの高さまで届くには{}メートルの紙が必要です".format(fold_num(Proxima_Centauri)))
```
